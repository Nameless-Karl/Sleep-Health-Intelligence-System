### 2. New & Transformed Variables
#### Systolic & Diastolic Columns:
Transformation: Split the Blood Pressure string (e.g., "120/80") into two integer columns.

Rationale: Blood pressure is a key indicator of cardiovascular stress, which significantly impacts sleep architecture; separate values allow for calculating Pulse Pressure or identifying hypertension stages.

#### Physical Activity Levels:
Transformation: Grouped Physical Activity Level into "Low," "Moderate," and "Active."
Rationale: High physical activity is correlated with deeper REM sleep, but excessive activity close to bedtime can be disruptive.
##### BMI-based Risk Categories:
Transformation: Simplified labels into "Underweight," "Normal Weight," "Overweight," and "Obese."
Rationale: High BMI is a primary clinical predictor for Sleep Apnea, making this variable essential for identifying high-risk sleep profiles.

#### How do BMI categories relate to sleep duration and sleep quality? (Assess whether overweight or obese groups exhibit poorer sleep patterns.)

```python
# Define the logical order for BMI
bmi_order = ['Underweight', 'Normal Weight', 'Overweight', 'Obese']
```

```python
# Convert the column to a Categorical type with this order
data['BMI Category'] = pd.Categorical(data['BMI Category'], categories=bmi_order, ordered=True)
```

```python
# Medians grouped by BMI Category
duration_medians = data.groupby('BMI Category', observed=False)['Sleep Duration'].median().sort_values()
quality_medians = data.groupby('BMI Category', observed=False)['Quality of Sleep'].median().sort_values()
```

```python
# Explicitly sort by index just to be safe
duration_medians = duration_medians.sort_index()
quality_medians = quality_medians.sort_index() # So the normal order of BMI's in the real world is reflected
```

```python
# The figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar Chart for Median Sleep Duration
duration_medians.plot(kind='bar', ax=ax1, color='skyblue', edgecolor='black')
ax1.set_title('Median Sleep Duration by BMI Category')
ax1.set_ylabel('Hours')
ax1.set_ylim(0, 10)  # Standardized scale for comparison
ax1.tick_params(axis='x', rotation=0)

# Bar Chart for Median Sleep Quality
quality_medians.plot(kind='bar', ax=ax2, color='salmon', edgecolor='black')
ax2.set_title('Median Sleep Quality by BMI Category')
ax2.set_ylabel('Quality Score (1-10)')
ax2.set_ylim(0, 10)
ax2.tick_params(axis='x', rotation=0)

# Titles
plt.suptitle('Sleep Metrics across BMI Categories', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

plt.tight_layout()
plt.show()
```

Systolic blood pressure is the top number in a blood pressure reading, representing the pressure in your arteries when your heart contracts and pushes blood out to the body, measuring the force against artery walls during a heartbeat.  It signifies the peak pressure as the heart pumps, while the lower number, diastolic pressure, reflects the pressure as the heart rests between beats. (Google)


Diastolic blood pressure is the bottom, lower number in a blood pressure reading, representing the pressure in the arteries when the heart muscle rests between beats and fills with blood. It measures the lowest pressure in the cardiovascular cycle, critical for coronary artery perfusion. (Google)

```python
# Filter Data for 'Overweight' Individuals ---
# We focus specifically on the group in question
overweight_data = data[data['BMI Category'] == 'Overweight']
```

```python
# Analyze the Impact of Stress on Sleep Quality ---
# Determine the dividing line for "High" vs "Low" stress in this group
median_stress = overweight_data['Stress Level'].median()
```

```python
# Split the group: Lower Stress (Better) vs. Higher Stress (Worse)
# Note: We want to show the benefit of being in the 'Low' group
low_stress_group = overweight_data[overweight_data['Stress Level'] < median_stress]
high_stress_group = overweight_data[overweight_data['Stress Level'] >= median_stress]
```

```python
# Calculate the average sleep quality for both groups
avg_sleep_low_stress = low_stress_group['Quality of Sleep'].mean()
avg_sleep_high_stress = high_stress_group['Quality of Sleep'].mean()
```

```python
# Calculate the percentage improvement gained by reducing stress
improvement = ((avg_sleep_low_stress - avg_sleep_high_stress) / avg_sleep_high_stress) * 100
```

```python
# Visualization
plt.figure(figsize=(10, 6))
# Plotting the comparison
bars = plt.bar(['High Stress\n(>= Median)', 'Low Stress\n(< Median)'], 
               [avg_sleep_high_stress, avg_sleep_low_stress], 
               color=['#ff9999', '#99ff99'], edgecolor='black')

plt.title(f'Impact of Stress Reduction on Sleep Quality\n(Overweight Individuals Only)', fontsize=14)
plt.ylabel('Average Sleep Quality (Scale 1-10)', fontsize=12)
plt.ylim(0, 10)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.2, round(yval, 2), ha='center', fontsize=12, fontweight='bold')

plt.show()
```

##### Biological Context:
Biochemically, high stress triggers cortisol release, which keeps the body in a catabolic 'fight or flight' state. For an 'efficient' metabolic system (as seen in this BMI category), this constant mobilization prevents the body from entering the restorative 'rest and digest' state required for high-quality sleep.

Additionally; One of the most efficient manners of stress management includes walking and as the Dataset suggests that walking more(higher steps taken daily) improves sleep quality. Stress level reduction acts as indicator for other physiological outcomes.
